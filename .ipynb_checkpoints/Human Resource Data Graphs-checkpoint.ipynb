```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import seaborn as sns
dataset = pd.read_csv('core_dataset.csv')
k1 = dataset.loc[:, ['Sex', 'Pay Rate', 'Department', 'State']]
```

```python
k1.head()
```

```python
hr = sns.barplot(x='State', y='Pay Rate', data=dataset,hue_order='Sex')
```

```python
kr = dataset.corr()
sns.heatmap(data=kr)
```

```python
hrp = dataset.pivot_table(index='State', columns='Department', values='Pay Rate', aggfunc=np.mean)
sns.heatmap(hrp, cmap='magma')
```

```python
hrp2 = dataset.pivot_table(index='State', columns=['Department', 'Sex'], values='Pay Rate', aggfunc=np.count_nonzero)
sns.heatmap(hrp2, cmap='magma')
```

```python
g = sns.FacetGrid(data=dataset,row='Sex',col='RaceDesc')
g.map(plt.scatter, 'Pay Rate', 'Department')
```

```python
sns.lmplot(x='Age',y='Pay Rate', data=dataset)
```

```python
sns.lmplot(x='Age',y='Pay Rate', data=dataset, hue='Sex',col='RaceDesc',row='Department',palette=dict(Male="b", FeMale="yellow"))
```
